```python
from pydantic import BaseModel
from typing_extensions import Optional
```

```python
class TestSubClass(BaseModel):
    k: int
    j: str

class TestClass(BaseModel):
    a: int
    b: str
    c: Optional[TestSubClass]
```

```python
json = {
    "a": 6,
    "b": "hello",
    "c": None
}
```

```python
model = TestClass(**json)
model
```
